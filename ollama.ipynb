### **Step 2: Initialize Server & Pull DeepHat**

```python
import subprocess
import os
import time

# Set host to listen on all interfaces
os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'

# Launch server as background process
subprocess.Popen(['ollama', 'serve'], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(5)  # Wait for boot
print("✅ Ollama server is active.")

# Pull the high-performance cybersecurity model
!ollama pull DeepHat/DeepHat-V1-7B
```
